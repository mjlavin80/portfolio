```python
import pandas as pd
df=pd.read_csv("song_metadata.csv", encoding="latin1")
from pandas import *
```

```python
titles = list(df["title"])
```

```python
years = list(df["year"])
```

```python
df["love"] = [1 if "love" in str(i).lower() else 0 for i in titles]
```

```python
df2=df[df.year != 0]
```

```python
loves = list(df["love"])
```

```python
df3 = df2[["title", "year", "love"]].copy()
df3 = df3.groupby(['year', 'love']).count()
df3 = df3.reset_index()
df3
```

df3

```python
number = list(df3["title"])
year = list(df3["year"]) 
```

```python
import plotly.plotly as py
import plotly.graph_objs as go
from plotly.graph_objs import *

# Create and style traces
trace0 = go.Scatter(
    x = year,
    y = number,
    line = dict(
        color = ('rgb(205, 12, 24)'),
        width = 4)
)
data = [trace0]

# Edit the layout
layout = dict(title = 'Usage of the Word \"Love\" in Song Titles',
              xaxis = dict(title = 'Year'),
              yaxis = dict(title = 'Number of Occurrences'),
              )

fig = dict(data=data, layout=layout)
py.iplot(fig, filename='styled-line')
```
